```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import math
import numpy as np
import random
from sklearn.datasets import make_blobs
import scipy
# from matplotlib import style
# style.use('ggplot')
```

```python
def get_key(val,my_dict):
    for key, value in my_dict.items():
         if (val[0] == value[0] and val[1] == value[1]):
            return key
```

```python
def calc_distance(X1, X2):
    return ((sum((X1 - X2)**2))**0.5)  #increasing spread here
```

```python
class K_Means:
    def __init__(self, k, tol=0.001, max_iter=30):
        self.k = k
        self.tol = tol
        self.max_iter = max_iter
    
    def closest_node(self,node, nodes):
        nodes = np.asarray(nodes)
        dist_2 = np.sum((nodes - node)**2, axis=1)
        return np.argmin(dist_2)

    def fit(self,data,snr_dic,clients):

        self.centroids = {}

        for i in range(self.k):
            self.centroids[i] = data[i]
#             print(self.centroids)

        for i in range(self.max_iter):
            self.classifications = {}

            for i in range(self.k):
                self.classifications[i] = []

            for featureset in data:
                distances = [np.linalg.norm(featureset-self.centroids[centroid]) for centroid in self.centroids]
                snrs=[]
                for c in self.centroids:
                    cl1=get_key(featureset,clients)
                    cl2=get_key(self.centroids[c],clients)
                    try:
                        try:
                            snr=snr_dic[tuple((tuple(featureset),tuple(self.centroids[c])))]
                        except:
                            snr=snr_dic[tuple((tuple(self.centroids[c]),tuple(featureset)))]
                    except:
                        snr=10000
                    snrs.append(snr)
                classification = snrs.index(max(snrs))
                self.classifications[classification].append(featureset)

            prev_centroids = dict(self.centroids)

            for classification in self.classifications:
                self.centroids[classification] = np.average(self.classifications[classification],axis=0)
            
            for c in range(len(self.centroids)):
                p=self.closest_node(self.centroids[c],data)
                self.centroids[c]=data[p]

            optimized = True

            for c in self.centroids:
                original_centroid = prev_centroids[c]
                current_centroid = self.centroids[c]
                if np.sum((current_centroid-original_centroid)/original_centroid*100.0) > self.tol:
#                     print(np.sum((current_centroid-original_centroid)/original_centroid*100.0))
                    optimized = False
            

            if optimized:
                break

    def predict(self,data):
        distances = [np.linalg.norm(data-self.centroids[centroid]) for centroid in self.centroids]
        classification = distances.index(min(distances))
        return classification
```

```python
def snr_list(clients):
    snr_dic={}
    csi_dic={}
    snr1_dic={}
    for i in range(len(clients)-1):
        for j in range(i+1,len(clients)):
            X1=clients['client'+str(i+1)]
            X2=clients['client'+str(j+1)]
            dis=calc_distance(X1,X2)
            path_loss=10*math.log10(1500/(dis*dis))#1000
            noise=random.uniform(0,5)
            snr_dic.update({tuple([tuple(X1),tuple(X2)]):path_loss-noise})
            csi_dic.update({tuple([tuple(X1),tuple(X2)]):random.random()})
#             snr1_dic.update({tuple(['client'+str(i+1),'client'+str(j+1)]):path_loss-noise})
    return(snr_dic,csi_dic)
```

```python
def getList(dict):
    list = []
    for key in dict.keys():
        list.append(key)
          
    return list
```

```python
def cluster_former(k,n):
    cluster_array, _ = make_blobs(n_samples=n,n_features=2, centers=k,random_state=56)
    # print(cluster_array)
    no=1
    clients={}
    for client in cluster_array:
        clients['client'+str(no)]=client
        no+=1
    clients_temp=clients
    snr_dic,csi_dic=snr_list(clients)
    model = K_Means(k=k)
    model.fit(cluster_array,snr_dic,clients)

    for centroid in model.centroids:
        plt.scatter(model.centroids[centroid][0], model.centroids[centroid][1],
                    marker="o", color="black",s=150)

    # no=0
    # for classification in model.classifications:
    # #     color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(k)]
    #     color=['red','blue','green','yellow']
    #     for featureset in model.classifications[classification]:
    #         plt.scatter(featureset[0], featureset[1], color=color[no])
    #     no+=1

    # plt.show()

    no=0
    cmap = plt.get_cmap('jet', 10)
    cmap=cmap.reversed()

    snrl=list(snr_dic[i] for i in snr_dic)

    intervals=np.linspace(min(snrl),max(snrl),10)
    #print(snrl)
    client_locs=getList(snr_dic)
    # print(type(client_locs))
    #print(client_locs[0][0])
    # print(len(model.classifications))
    for classification in model.classifications:
        color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(k)]
       #color=['red','blue','green','yellow']
        #cx=next(cycol)
        for featureset in model.classifications[classification]:

            plt.scatter(featureset[0], featureset[1], c=color[classification])
            #index=np.searchsorted(intervals,snr)
            #print(model.centroids[classification][0], model.centroids[classification][1])
            clients=[featureset[0], featureset[1]]
            client_t=tuple(clients)
            centroid_cl=[model.centroids[classification][0], model.centroids[classification][1]]
            centroid_t=tuple(centroid_cl)
            l=[client_t,centroid_t]
            l_rev=[centroid_t,client_t]
            #print(centroid_cl)
            xs=[clients[0],centroid_cl[0]]
            ys=[clients[1],centroid_cl[1]]
            line=tuple(l)
            line_rev=tuple(l_rev)
    #         print(line)
            if(line[0][0]==line[1][0] and line[0][1]==line[1][1]):
                continue
            else:

                if(line in snr_dic):
                    snr=snr_dic[line]
                else:
                    snr=snr_dic[line_rev]

                index=np.searchsorted(intervals,snr)
                plt.plot(xs,ys,color=cmap(index),linewidth=1.5)

        no+=1

    norm = matplotlib.colors.Normalize(vmin=min(snrl), vmax=max(snrl))
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar=plt.colorbar(sm, ticks=np.linspace(min(snrl), max(snrl), 10))
    cbar.set_label("Signal-to-Noise Ratio")         
    plt.show()
    clients=clients_temp
    
    
    cluster_list=[]
    heads_list=[]
    for classification in model.classifications:
        members=[]
        snrs=[]
        csis=[]
        cluster_head=get_key(model.centroids[classification],clients)
        heads_list.append(model.centroids[classification])
        cluster_dic={}
        for m in model.classifications[classification]:
            member=get_key([m[0],m[1]],clients)
            if(m[0]!=model.centroids[classification][0] and m[1]!=model.centroids[classification][1]):
                members.append(member)
            try:
                try:
                    snr=snr_dic[tuple([tuple([m[0],m[1]]),tuple(model.centroids[classification])])]
                    snrs.append(snr)
                except:
                    snr=snr_dic[tuple([tuple(model.centroids[classification]),tuple([m[0],m[1]])])]
                    snrs.append(snr)
            except:
                None

            try:
                try:
                    csi=csi_dic[tuple([tuple([m[0],m[1]]),tuple(model.centroids[classification])])]
                    csis.append(csi)
                except:
                    csi=csi_dic[tuple([tuple(model.centroids[classification]),tuple([m[0],m[1]])])]
                    csis.append(csi)
            except:
                None

        cluster_dic.update({'Cluster Head':cluster_head})
        cluster_dic.update({'Members':members})
        cluster_dic.update({'SNR':snrs})
        cluster_dic.update({'CSI':csis})
        cluster_list.append(cluster_dic)
        
        head_snr_list=[]
#         print(heads_list)
        for h1 in range(len(heads_list)-1):
            for h2 in range(h1+1,len(heads_list)):
                try:
                    snr2=snr_dic[tuple([tuple(heads_list[h1]),tuple(heads_list[h2])])]
                except:
                    snr2=snr_dic[tuple([tuple(heads_list[h2]),tuple(heads_list[h1])])]
                head_snr_list.append([get_key(heads_list[h1],clients),get_key(heads_list[h2],clients),snr2])
    
#     print(head_snr_list)
                
    
    return(cluster_list,head_snr_list)
    
```
